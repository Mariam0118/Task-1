```python
class Book :
    def __init__(self, title, author, isbn, available = True):
        self.title = title
        self.author = author
        self.__isbn = isbn
        self.available = available

    def display_info (self) :
        availability = "Available" if self.available else "Not Available"
        print(f"Title: {self.title} , Author: {self.author} , ISBN: {self.__isbn} , Status: {availability}")

    def get_isbn (self) :
        return self.__isbn

    def set_isbn(self, new_isbn):
        if isinstance(new_isbn, str) and new_isbn.strip():
            self.__isbn = new_isbn
        else:
            print("Invalid ISBN. Must be a non-empty string.")

class Member :
    def __init__(self, name , membership_id) :
        self.name = name
        self.__membership_id = membership_id 
        self.borrowed_books = []


    def get_membership_id (self) :
        return self.__membership_id

    def set_membership_id (self , new_id) :
        if isinstance(new_id, str) and new_id.strip() :
            self.__membership_id = new_id
        else:
            print("Invalid Membership ID. Must be a non-empty string.")

    def borrow_book (self , book) :
        if book.available :
            book.available = False
            self.borrowed_books.append(book)
            print(f"{self.name} borrowed '{book.title}'.")
        else:
            print(f"Sorry, '{book.title}' is currently not available.")

    def return_book(self , book) :
        if book in self.borrowed_books :
            book.available = True
            self.borrowed_books.remove(book)
            print(f"{self.name} returned {book.title} ")
        else:
            print(f"{self.name} has not borrowed {book.title}")


class StaffMember (Member) :
    def __init__(self , name , membership_id , staff_id ) :
        super().__init__(name , membership_id)
        self.staff_id = staff_id

    def add_book (self , library , title , author , isbn):
        new_book = Book(title, author, isbn)
        library.books.append(new_book)
        print(f"Staff member {self.name} added '{title}' to the library.")

class Library:
    def __init__(self):
        self.books = []

    def display_books(self):
        if not self.books:
            print("Library has no books.")
        else:
            for book in self.books:
                book.display_info()
```
